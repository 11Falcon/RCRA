```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def generate_paragraph(description, model_name="gpt2", max_length=150):
    """
    Generate a well-structured paragraph from a given description using GPT-2.

    Args:
        description (str): The input description.
        model_name (str): The name of the GPT-2 model to use (default: "gpt2").
        max_length (int): The maximum length of the generated text (default: 150).

    Returns:
        str: The generated paragraph.
    """
    # Load the GPT-2 model and tokenizer
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)
    model = GPT2LMHeadModel.from_pretrained(model_name)

    # Encode the input description
    input_ids = tokenizer.encode(description, return_tensors="pt")

    # Generate text
    output = model.generate(
        input_ids,
        max_length=max_length,
        num_return_sequences=1,
        no_repeat_ngram_size=2,
        top_k=50,
        top_p=0.95,
        temperature=1.0,
        pad_token_id=tokenizer.eos_token_id
    )

    # Decode the generated text
    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)

    return generated_text

# Example usage
if __name__ == "__main__":
    description = "A car is a road vehicle, typically with four wheels, powered by an internal combustion engine or an electric motor."
    paragraph = generate_paragraph(description)
    print("Generated Paragraph:")
    print(paragraph)
```
